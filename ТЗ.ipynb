```python
# импортируем необходимые библиотеки
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# загрузим и посмотрим на имеющиеся данные
data = pd.read_csv('test_data.csv')
data.head(10)
```

```python
# посмотрим на характеристики и тип данных
data.info()
data.describe()
```

```python
# изменим тип данных у столбцов с денежным выражением на float
columns = ['Spend', 'Revenue_ad1d', 'Revenue_in_app1d', 'Revenue_ad7d', 'Revenue_in_app7d', 'Revenue_ad30d', 'Revenue_in_app30d']
data[columns]=data[columns].replace(',', '.', regex=True).astype(float)
```

```python
data.info()
```

### Рассчитаем маркетинговые метрики в разрезе кампаний

```python
# CTR (Сlick-through rate - коэффициент кликабельности)
data['CTR'] = data['Clicks']/data['Impressions'] * 100

# Conversion Rate (коэффициент конверсии, регистрации)
data['Conversion Rate'] = data['Registrations']/data['Installs'] * 100

# CPI (Cost Per Install - стоимость установки)
data['CPI'] = data['Spend']/data['Installs']

# CPC (Cost Per Click - стоимость клика)
data['CPC'] = data['Spend']/data['Clicks']

# CPR (Cost per Registration - стоимость регистрации)
data['CPR'] = data['Spend']/ data['Registrations']

# Paying User Rate (коэффициент платящих пользователей)
data['Paying User Rate'] = data['Payers30d']/data['Registrations'] * 100

# ROI (Return on Investment - возврат инвестиций)
data['ROI'] = (data['Revenue_in_app30d'] + data['Revenue_ad30d'] - data['Spend']) / data['Spend'] * 100

# ARPPU (Average Revenue Per Paying User - средний доход на одного платящего пользователя) за месяц
data['ARPPU 30d'] = data['Revenue_in_app30d']/data['Payers30d']
```
